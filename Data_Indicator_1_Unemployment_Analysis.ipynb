# Data Cleaning


1.   Removing empty columns
2.   Removing missing values
3. Convert percentage columns from object to float64



### 1. Removing empty columns

```python
df_dropped_columns = df.drop(['Unnamed: 11',
                              'Unnamed: 12',
                              'Unnamed: 13',
                              'Unnamed: 14',
                              'Unnamed: 15',
                              'Unnamed: 16',
                              'Unnamed: 17',
                              'Unnamed: 18',
                              'Unnamed: 19',
                              'Unnamed: 20',
                              'Unnamed: 21',
                              'Column2'], axis=1)

# better logic suggested by ChatGPT is -- df_dropped_columns = df.drop(columns=[f'Unnamed: {i}' for i in range(11, 22)], errors='ignore')
df_dropped_columns.info()
```

### 2. Removing missing values

```python
df_final = df_dropped_columns.dropna()
df_final.info()
```

Removing unnecessary columns and missing records have reduced our dataset to 10 columns and 101 rows.

### 3. Convert percentage columns from object to float64

```python
df_final.loc[:, 'National\nUnemployment Rate SA (%)'] = df_final['National\nUnemployment Rate SA (%)'].replace('%', '', regex=True).astype(float)
df_final.loc[:, 'Dublin\nUnemployment Rate SA (%)'] = df_final['Dublin\nUnemployment Rate SA (%)'].replace('%', '', regex=True).astype(float)
df_final.loc[:, 'Dublin - National\nUnemployment Rate SA (%)'] = df_final['Dublin - National\nUnemployment Rate SA (%)'].replace('%', '', regex=True).astype(float)
```

# Data Visualisation


1.   Number of people employed in Dublin since 2018
2.   People employed vs. unemployed in Dublin since 2018
3. Percentage Comparison of Dublin and National Unemployment Rates
4. Dublin’s unemployment rate and the national unemployment rate difference




```python
df_final

# 1   Quarter – This likely represents the time period (e.g., Q1 2023, Q2 2023) for the data.
# 2   National Unemployment Rate SA (%)  – The seasonally adjusted (SA) unemployment rate for the entire country.
# 3  Dublin Unemployed SA (000) – The number of unemployed people in Dublin, in thousands, with seasonal adjustments.
# 4  Dublin Unemployment Rate SA (%) – The percentage of the labor force in Dublin that is unemployed (seasonally adjusted).
# 5   Dublin Employed SA (000) – The number of employed people in Dublin, in thousands, with seasonal adjustments.
# 6   QoQ Dublin Employed – Quarter-on-quarter change in the number of employed people in Dublin (likely a percentage or absolute change).
# 7   QoQ Dublin Unemployment Rate – Quarter-on-quarter change in the unemployment rate for Dublin.
# 8   YoY Dublin Employed  – Year-on-year change in employment numbers for Dublin.
# 9   %YoY Dublin Unemployed – Year-on-year percentage change in unemployment figures for Dublin.
# 10  Dublin - National Unemployment Rate SA (%) – The difference between Dublin’s unemployment rate and the national unemployment rate.
```

```python
df_final.columns
```

### 1. Number of people employed in Dublin since 2018

```python
# plt.plot(df_final['Quarter'], df_final['Dublin\nEmployed SA (000)'])

plt.figure(figsize=(20, 5)) # Width x Height

plt.plot(df_final['Quarter'][-27:], df_final['Dublin\nEmployed SA (000)'][-27:])

plt.title("People employed in Dublin since 2018")
plt.xlabel("Year")
plt.ylabel("People \n (In Thousands)")

plt.show()
```

### 2. People employed vs. unemployed in Dublin since 1998

```python
# data
x = df_final['Quarter'][-27:]
y1 = df_final['Dublin\nEmployed SA (000)'][-27:]
y2 = df_final['Dublin\nUnemployed SA (000)'][-27:]


plt.figure(figsize=(20, 5)) # Width x Height

# Plot lines with labels
plt.plot(x, y1, linestyle='-', color='blue', label='Employed')
plt.plot(x, y2, linestyle='--', color='red', label='Unemployed')

plt.title("Employed vs unemployed trend since 2018")
plt.xlabel("Year")
plt.ylabel("People \n (In Thousands)")


plt.legend()
plt.show()
```

### 3. Percentage Comparison of Dublin and National Unemployment Rates

```python
# data
quarters = df_final['Quarter'][-27:] # only take years from Q1 2018 till end
national_employment_rate = df_final['National\nUnemployment Rate SA (%)'][-27:]
dublin_unemployment_rate = df_final['Dublin\nUnemployment Rate SA (%)'][-27:]

n = len(quarters)
x = np.arange(n)
bar_width = 0.4

plt.figure(figsize=(20, 5)) # Width x Height

# Create grouped bar chart
plt.bar(x - bar_width/2, national_employment_rate, width=bar_width, label='National Unemployment Rate', color='orange')
plt.bar(x + bar_width/2, dublin_unemployment_rate, width=bar_width, label='Dublin Unemployment Rate', color='green')

# Customize x-axis
plt.xticks(x, quarters)  # Set x-ticks to category labels
plt.xlabel('Quarters')
plt.ylabel('Percentage')
plt.title('% Comparison of Dublin and National Unemployment')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

### 4. Dublin’s unemployment rate and the national unemployment rate difference

```python
# data
quarters = df_final['Quarter'][-27:] # only take years from Q1 2018 till end
unemployment_difference = df_final['Dublin - National\nUnemployment Rate SA (%)'][-27:]

plt.figure(figsize=(20, 5)) # Width x Height
plt.xlabel('Quarters')
plt.ylabel('Percentage')
plt.title('Dublin and national unemployment rate difference')
plt.bar(quarters, unemployment_difference, color='orange', label="Unemployment Rate Difference")
plt.show()
```
