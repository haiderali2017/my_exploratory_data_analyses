# Data Cleaning
1. Removing missing values
3. Renaming columns
4. Removing character from records

### 1. Removing missing values

In this step, we are removing missing values.

```python
df = df.dropna().reset_index(drop=True)
```

### 2. Renaming columns

In this step, we will rename few columns for better understanding.

```python
cols = df.columns
df = df.rename(columns=({
    cols[1]:'Passengers count',
    cols[2]: 'Passengers (In millions)',
    cols[7]: 'Changes in previous four quarters',
    cols[8]: 'Sum of previous four quarters'
}))
```

### 3. Removing characters from records

In this step, we are removing % and , from records to help create sensible visualisations.

```python
df['Passengers (In millions)'] = df['Passengers (In millions)'].str.replace(',', '.').astype(float)
df['Changes in previous four quarters'] = df['Changes in previous four quarters'].str.replace(',', '').astype(float)
df['Sum of previous four quarters'] = df['Sum of previous four quarters'].str.replace(',', '').astype(float)
df['% YoY'] = df['% YoY'].str.replace('%', '').astype(float)
df['%QoQ'] = df['%QoQ'].str.replace('%', '').astype(float)
df['YoY'] = df['YoY'].str.replace('%', '').str.replace(',', '').astype(float)
df['% Average YoY'] = df['% Average YoY'].str.replace('%', '').astype(float)

df['Passengers (In millions)'] = df['Passengers (In millions)'].round(1)

# def replace_characters(df, col_name, replace_char, replace_with):
#   df[col_name] = df[col_name].replace(replace_char, replace_with, regex=True).astype('float')

# columns_ = df.columns # ['Quarter', 'Passengers count', 'Passengers (In millions)', '% YoY', '%QoQ', 'YoY', '% Average YoY', 'Changes in previous four quarters','Sum of previous four quarters']

# for col in columns_:
#   if col in ['% YoY', '%QoQ', 'YoY']:
#     replace_characters(df, col, '%', '')
#   elif col in ['YoY', 'Sum of previous four quarters', 'Changes in previous four quarters']:
#     replace_characters(df, col, ',', '')
#   elif col == 'Passengers (In millions)':
#     replace_characters(df, col, ',', '.')
```

# Data Visualisation

### Line Chart - Dublin Airport passenger arrivals

```python
fig = px.line(df, x='Quarter', y=df['Passengers (In millions)'], title='Dublin Airport passenger arrivals')

# Update the x-axis to show all years
fig.update_xaxes(
    tickmode='array',  # Set tick mode to 'array'
    tickvals=df['Quarter'],  # Specify the tick values (all quarter)
    ticktext=df['Quarter']   # Specify the tick labels (all quarter)
)

fig.show()
```

### Line Chart - % YoY Dublin Airport passenger arrivals

```python
fig = px.line(df, x='Quarter', y=df['% YoY'], title='Year-on-Year % Dublin Airport passenger arrivals')

# Update the x-axis to show all years
fig.update_xaxes(
    tickmode='array',  # Set tick mode to 'array'
    tickvals=df['Quarter'],  # Specify the tick values (all quarter)
    ticktext=df['Quarter']   # Specify the tick labels (all quarter)
)

fig.show()
```

### Line Chart - % QoQ Dublin Airport passenger arrivals

```python
fig = px.line(df, x='Quarter', y=df['%QoQ'], title='Quarter-on-Quarter % Dublin Airport passenger arrivals')

# Update the x-axis to show all years
fig.update_xaxes(
    tickmode='array',  # Set tick mode to 'array'
    tickvals=df['Quarter'],  # Specify the tick values (all quarter)
    ticktext=df['Quarter']   # Specify the tick labels (all quarter)
)

fig.show()
```

### Pie Chart - 2020 Quarter distribution of passenger arrivals

```python
# filter rows in a DataFrame
filtered_df = df.loc[df['Quarter'].str.contains('20'),['Quarter','Passengers (In millions)']]

fig = px.pie(filtered_df, values='Passengers (In millions)', names='Quarter', color_discrete_sequence=px.colors.qualitative.Bold)
fig.show()
```

```python
# filter rows in a DataFrame
q1 = df.loc[df['Quarter'].str.contains('Q1'),'Passengers (In millions)']
q2 = df.loc[df['Quarter'].str.contains('Q2'),'Passengers (In millions)']
q3 = df.loc[df['Quarter'].str.contains('Q3'),'Passengers (In millions)']
q4 = df.loc[df['Quarter'].str.contains('Q4'),'Passengers (In millions)']

quarters = pd.DataFrame({
    'Quarters': ['Q1', 'Q2', 'Q3', 'Q4'],
    'Values': [q1.sum(), q2.sum(), q3.sum(), q4.sum()]
})

fig = px.bar(quarters, x='Quarters', y='Values', width=600, color='Values')
fig.show()
```
