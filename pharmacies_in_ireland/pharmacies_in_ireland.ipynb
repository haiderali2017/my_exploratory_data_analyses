# Exploratory Data Analysis



##1.   Number of pharmacies in all counties



```python
# Compute the count of each category and sort in descending order
counts = df['County'].value_counts().sort_values(ascending=False)

# Create the countplot with horizontal orientation and sorted categories
ax = sns.countplot(y='County', data=df, order=counts.index)


# Add count labels to the bars
for p in ax.patches:
    count = int(p.get_width())  # Convert the count to an integer
    ax.annotate(f'{count}', (count, p.get_y() + p.get_height() / 2), ha='left', va='center')


# In this code, after creating the countplot, we iterate over each bar using ax.patches. 
# For each bar, we use the annotate() function to add a text label indicating the count. 
# The p.get_width() gives the count value of the bar, and we position the label at the center of the bar using p.get_width() and p.get_y() + p.get_height() / 2. 

# Set labels and title
plt.xlabel('Count')
plt.ylabel('Counties')
plt.title('Count of Pharmacies')

plt.show()
```

*It is evident from above visualisation that Dublin ranks first with the most pharmacies in Ireland and Leitrim with the least.*





##2. Pharmaceutical companies  



```python
# Calculate value counts
value_counts = df['Service name'].value_counts()[:10]

# Plot the occurrences
plt.barh(value_counts.index[::-1], value_counts.values[::-1])

# Add labels to the bars
for i, v in enumerate(value_counts.values[::-1]):
    plt.text(v, i, str(v), ha='left', va='center')

# The plt.text() function is used inside a loop to iterate over the values (v) and their corresponding indices (i). 
# It places the value as a label next to each bar. The ha='left' parameter aligns the text to the left of the bar, 
# and va='center' centers it vertically within the bar.

# Set labels and title
plt.xlabel('Count')
plt.ylabel('Pharmacies')
plt.title('Occurrences of Pharmacies')

# Display the plot
plt.show()
```

*It is evident from above visualisation that Boots run the most pharmeutical branches in Ireland.*





##3. Top 5 pharmacies in Dublin 



```python
County = 'Dublin'

# Calculate value counts
dublin_pharm_df = df[df['County']==County]['Service name']
value_counts = dublin_pharm_df.value_counts()[:5]

# Plot the occurrences
plt.barh(value_counts.index[::-1], value_counts.values[::-1])

# Add labels to the bars
for i, v in enumerate(value_counts.values[::-1]):
    plt.text(v, i, str(v), ha='left', va='center')

# The plt.text() function is used inside a loop to iterate over the values (v) and their corresponding indices (i). 
# It places the value as a label next to each bar. The ha='left' parameter aligns the text to the left of the bar, 
# and va='center' centers it vertically within the bar.

# Set labels and title
plt.xlabel('Count')
plt.ylabel('Pharmacies')
plt.title('Top 5 Pharmacies in '+County)

# Display the plot
plt.show()
```

*Number of pharmacies in individual counties may be different than whole Ireland. It is evident from above visualisation that DocMorris Pharmacy run the most branches in Dublin, even more than Boots which is the leader in whole Ireland.*
